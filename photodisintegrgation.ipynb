Photodisintegration with non-extensive statistics

```python
# The theory for the nonextensive planck distribution is given in the following paper:
# Constantino Tsallis. Introduction to nonextensive statistical mechanics: approaching a complex world. Springer Science & Business Media, 2009.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
q = 1.075
x = np.linspace(0, 10, 1000)
```

```python
y = ((1-np.exp(-x))**(q-1))*(1+(1-q)*x*((1+np.exp(-x))/(1-np.exp(-x)) - (x*(1+3*np.exp(-x)))/(2*(1-np.exp(-x))*(1-np.exp(-x)))))
y2 = x*x/((-1+np.exp(x)))
y1 = y*y2
q=0.95
y = y2*((1-np.exp(-x))**(q-1))*(1+(1-q)*x*((1+np.exp(-x))/(1-np.exp(-x)) - (x*(1+3*np.exp(-x)))/(2*(1-np.exp(-x))*(1-np.exp(-x)))))
```

```python
plt.plot(x,y1,label='q=1.075')
plt.plot(x,y2,label='q=1')
plt.plot(x,y,label='q=0.95')
plt.legend(loc='upper right')
plt.xlabel('$E/k_{B}T$')
plt.title('Planck Distribution for different values of q')
```
